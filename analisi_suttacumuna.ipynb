```python
from geopandas import read_file, GeoDataFrame
from pandas import concat, read_csv, NA
import matplotlib.pyplot as plt
```

```python
gdf1 = read_file("ASC_21/ASC_Liv_1_WGS84.shp")
A = gdf1[gdf1["COD_REG"] == 19]

gdf2 = read_file("ASC_21/ASC_Liv_2_WGS84.shp")
B = gdf2[gdf2["COD_REG"] == 19]

gdf3 = read_file("ASC_21/ASC_Liv_3_WGS84.shp")
C = gdf3[gdf3["COD_REG"] == 19]
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

A.plot(ax=ax, color='lightgray', edgecolor='black')
B.plot(ax=ax, color='blue', edgecolor='black')
C.plot(ax=ax, color='red', edgecolor='black')
plt.title("Municipalities in Italy")
plt.show()
```
